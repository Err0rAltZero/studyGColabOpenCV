```
from google.colab import output
import cv2 as cv
import numpy as np 
from matplotlib import pyplot as plt
import math as m
#NecessÃ¡rios para os widgets
import ipywidgets as widgets
from ipywidgets import HBox, VBox
```

```
img=cv.imread("/content/drive/My Drive/Colab Files/imgMy/fruits.jpg",0)
plt.imshow(img,"gray")
plt.title("Imagem original em grayscale")
plt.show()
```

```
#InicializaÃ§Ã£o dos Widgets
caption = widgets.Label(value='Escolher o valor de limiar (T):')

sliderT = widgets.IntSlider(min=0, max=255, value=175, description='Threshold')
dropT= widgets.Dropdown(
    options=[("cv.THRESH_BINARY",0),("cv.THRESH_BINARY_INV",1),("cv.THRESH_TRUNC",2),("cv.THRESH_TOZERO",3),("cv.THRESH_TOZERO_INV",4)],
    value=0,
    description='MÃ©todo:',
    disabled=False,
)
buttonT = widgets.Button(description="Limiarizar")
buttonC = widgets.Button(description="Limpar")
#Layout dos Widgets
hBox1 = HBox([sliderT, dropT])
hBox2 = HBox([buttonT, buttonC])
display(caption, VBox([hBox1, hBox2]))
#Vetor para TÃ­tulos
methods = ["cv.THRESH_BINARY", "cv.THRESH_BINARY_INV", "cv.THRESH_TRUNC", "cv.THRESH_TOZERO", "cv.THRESH_TOZERO_INV"]

#Obter os parametros, parametrizar a limiarizaÃ§Ã£o e exibir as figuras
def on_buttonT_clicked(b):
    ret,imgThresh = cv.threshold(img,sliderT.value,255,dropT.value)      
    plotT= plt.imshow(imgThresh,"gray")
    plt.title("T= "+str(sliderT.value)+" MÃ©todo: "+methods[dropT.value])    
    plt.show()
#Limpar as figuras
def on_buttonC_clicked(b):
    output.clear() #Limpa todas as saÃ­das
    display(caption, VBox([hBox1, hBox2])) #Exibe novamente os Widgets

#AÃ§Ãµes dos botÃµes    
buttonT.on_click(on_buttonT_clicked)
buttonC.on_click(on_buttonC_clicked)
```
